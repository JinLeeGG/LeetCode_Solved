```python
class Solution(object):
    def romanToInt(self, s):
        """
        :type s: str
        :rtype: int
        """
        Roman = {"I" : 1, "V" : 5, "X" : 10, "L" : 50, "C" : 100, "D" : 500, "M" : 1000}
        count = 0

        # check for every string
        for i in range(len(s)):

            # if next string exists and current string is smaller than next string
            if i + 1 < len(s) and Roman[s[i]] < Roman[s[i + 1]]:
              # subtract current string value
              count -= Roman[s[i]]
              # Otherwise, add current String value
            else:
              count += Roman[s[i]]

        return count





s = Solution()
s.romanToInt("MCMXCIV")

# 로마 숫자는 현재 문자의 값이 다음 문자의 값보다 작으면 빼고, 그렇지 않으면 더하는 방식으로 계산됩니다.
```
